```python
import pandas as pd
input_data = pd.read_csv("diabetes.tab.txt", sep='\t', header=[0])
```

```python
# print(input_data.mean())
input_data_std = input_data.std(0)*3
upper_input_data = input_data.mean() + input_data.std(0) * 3
lower_input_data = input_data.mean() - input_data.std(0) * 3

# print(upper_input_data)
# print(lower_input_data)

input_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

diabetes_data = input_data.drop(['Y'], axis=1)
print(diabetes_data)
diabetes_target =input_data['Y']

X_train, X_test, y_train, y_test = train_test_split(diabetes_data, diabetes_target, test_size=0.4, random_state=0)
print(X_train.shape)
print(X_test.shape)

# model_diabetes= LinearRegression().fit(diabetes_data, diabetes_target)
model_diabetes= LinearRegression().fit(X_train, X_test) #위에 식보다 부정확해질 수 있음 테스트 케이스가 적어져서
```

```python
model_diabetes.coef_
```

```python
import matplotlib.pylab as plt
predicitors = model_diabetes.predict(X_train)
plt.scatter(y_train, predicitors)
plt.xlabel("")
plt.ylabel("")
plt.title("")
plt.show()
```
